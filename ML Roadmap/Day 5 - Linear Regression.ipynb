```python
import pandas as pd
df=pd.read_csv('dataset/student_scores.csv')
df
```

```python
len(df)
```

```python
X=df['Hours']
y=df['Scores']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
x_train.shape
```

```python
type(x_train)
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
#build model
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr
```

```python
#train 
#code - fit
x_train_twodi=x_train.values.reshape(-1,1)
lr.fit(x_train_twodi,y_train) 
```

```python
lr.coef_ #value of w
```

```python
lr.intercept_  #b - bias
```

```python
x=4.5
w=9.53227213
b=2.9760987546639583
y=w*x+b
print(y)
```

```python
lr.predict([[4.5]])
```

```python
y_train
```

```python
pred_value=lr.predict(x_test.values.reshape(-1,1))
pred_value
```

```python
y_test
```

```python
for x,y in zip(pred_value,y_test):
    print(f"pred: {int(x)} act: {y}")
```

```python
#evaluate
from sklearn.metrics import mean_absolute_error, mean_squared_error
mae=mean_absolute_error(y_test,pred_value)
print(f"Mean absolute error {mae}")
```

```python
mse=mean_squared_error(y_test,pred_value)
print(f"Mean squared error {mse}")
```

```python
#training data testing
#x_train,y_train

train_pred_value=lr.predict(x_train.values.reshape(-1,1)).round()
train_pred_value
```

```python
mae=mean_absolute_error(y_train,train_pred_value)
print(f"Mean absolute error {mae}")  #4.756069196104083
```

```python
mse=mean_squared_error(y_train,train_pred_value)
print(f"Mean squared error {mse}")
```
