```python
import pandas as pd
df=pd.read_csv('dataset/Sales_Data/alldf.csv')
df
```

```python
df['Sales Amount'].sum()
```

```python
df['Product'].nunique()
```

```python
month='November'
city='Boston'
product=['Google Phone','Lightning Charging Cable']

df_selection=df.query("Product==@product and Month==@month and City==@city")
df_selection
```

```python
df_selection.groupby('Product')['Sales Amount'].sum().sort_values(ascending=False)
```

```python
sales_by_month=df_selection.groupby('Month')['Sales Amount'].sum().sort_values(ascending=False)
sales_by_month
```
